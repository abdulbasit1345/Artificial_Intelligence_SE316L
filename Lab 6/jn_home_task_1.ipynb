```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
file = pd.read_excel('marketing_campaign.csv', sep='\t')
file.head()
```

```python
file['Education'] = file['Education'].str.strip()
file.dropna(axis=0, subset=['Marital_Status'], inplace=True)
file.dropna(axis=0, subset=['Education'], inplace=True)
file.dropna(axis=0, subset=['Income'], inplace=True)

#file['Martal_Status'] = file['Martal_Status'].astype('str')
#file = file[~file['Martal_Status'].str.contains('C')]
file
```

```python
basket = (file
.sum().unstack().reset_index().fillna(0).setIndex('ID'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0 
    if x >= 1:
        return 1

```

```python
frequent_itemsets = apriori(file, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, file, metric="lift", min_threshold=1)
rules.head()
```
